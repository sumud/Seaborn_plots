```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset("tips")
data.head()
```

```python
#fit regression model
#lmplot combines regplot with facetgrid to show multiple fits
g1 = sns.lmplot(x='total_bill', y = 'tip', hue='sex',data= data)
plt.xlabel("Total Bill($)")
plt.ylabel("Tip($)")
plt.title("Tip amount vs Total bill amount")
```

```python
import pandas as pd
import numpy as np
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# create subplot for plotiing conditional relationships
gr= sns.FacetGrid(titanic, col = 'survived', row = 'sex')
#now draw a histogram showing age onto the facetgrid
gr = gr.map(plt.hist, 'age')
```

```python
# draw a categorical plot onto facetgrid
sns.factorplot(x='pclass', y='survived', hue='sex', data=titanic)
```

```python
# categorical plots
#scatterplot with one categorical variable
sns.stripplot(data=titanic, x='sex', y='age')
```

```python
#categorical plot with non overlapping points
sns.swarmplot(data= titanic, x= 'sex', y = 'age')
```

```python
#bar charts
sns.barplot(data=titanic, x='sex',y='survived', hue='class')
#showing point estimates, CI 
```

```python
#count plot- showing count of a variable
sns.countplot(data=titanic, x='embark_town', palette='husl')
```

```python
#boxplot 
sns.boxplot(data=titanic, x='alive', y = 'age', hue='adult_male')
```

```python
# violin plot
sns.violinplot(data=titanic, x="sex", y="age", hue="survived")
```

```python
#pairwise plotting relationship
h=sns.pairplot(titanic)
```

```python
#plot bivariate distribution
sns.jointplot(data=titanic, x='age', y='pclass', kind='kde')
```

```python
#linear regression fit
#regplot shows only one relationship
sns.regplot(data=titanic, x='age',y='fare')
```
